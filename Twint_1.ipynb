```python
# twint Experiment
import pandas as pd

import twint
```

```python
pd.set_option('display.max_columns', None)
```

```python
import nest_asyncio
nest_asyncio.apply()
import time

import os
import re
timestr = time.strftime("%Y%m%d")

c = twint.Config()
c.Limit = 2000
c.Lang = "en"
c.Store_csv = True
c.Search = "#bitcoin"
c.Output = timestr + "_en_btc.csv"
twint.run.Search(c)
```

```python
df = pd.read_csv('20210729_en_btc.csv')
df.columns
```

```python
df.head(2)
```

```python
len(df)
```

```python
df_tweets = pd.DataFrame(df['tweet'])
df_tweets
```

```python
# cleaning steps
def cleanTwT(twt):
    twt = re.sub('#bitcoin', 'bitcoin', twt)
    twt = re.sub('#Bitcoin', 'Bitcoin', twt)
    twt = re.sub('#[A-Za-z0-9]+', '', twt)
    twt = re.sub('@[A-Za-z0-9]+', '', twt)
    twt = re.sub('//n', '', twt)
    twt = re.sub('https?:\/\/\S+', '', twt)
    return twt
```

```python
df_tweets['clean_tweet'] = df_tweets['tweet'].apply(cleanTwT)
df_tweets.head()
```

```python
from textblob import TextBlob
```

```python

# get subjectivity
def getSubjectivity(twt):
    return TextBlob(twt).sentiment.subjectivity

# polarity
def getPolarity(twt):
    return TextBlob(twt).sentiment.polarity

df_tweets['subjectivity'] = df_tweets['clean_tweet'].apply(getSubjectivity)
df_tweets['polarity'] = df_tweets['clean_tweet'].apply(getPolarity)

df_tweets.head()
```

```python
# get sentiment text
def getSentiment(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
```

```python
# insert into df
df_tweets['sentiment'] = df_tweets['polarity'].apply(getSentiment)
df_tweets.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
for i in range(0, df_tweets.shape[0]):
    plt.scatter(df_tweets['polarity'][i], df_tweets['subjectivity'][i], color='Purple')
plt.title('polarity vs. subjectivity - Bitcoin')
plt.ylabel('subjectivity')
plt.xlabel('polarity')
plt.show()
```

```python
df_tweets['sentiment'].value_counts().plot(kind='bar')

plt.title('Sentiment Barplot - Bitcoin')
plt.ylabel('# of tweets')
plt.xlabel('sentiment')
plt.show()
```
